### 2. Введение в генеративно-состязательные сети

**Генеративно-состязательные сети, или GAN**, представляют собой генеративную модель, основанную на глубоком обучении.

Архитектура GAN была впервые описана в статье «Генеративные состязательные сети» Иэна Гудфеллоу и др.

Стандартизованный подход под названием Deep Convolutional Generative Adversarial Networks, или DCGAN, который привёл к появлению более стабильных моделей, позже был формализован Алеком Редфордом и др. в статье «Обучение представлению без учителя с помощью глубоких свёрточных порождающих состязательных сетей».

**Архитектура модели GAN включает две подмодели:**

- модель генератора для создания новых экземпляров данных;
- модель дискриминатора для классификации сгенерированных моделью генератора примеров на предмет их реальности или поддельности.

Архитектура GAN

В отличие от VAE, который явно изучает распределение правдоподобия  через функцию потерь, GAN явно не изучает распределение вероятности. Генераторы GAN служат для генерации изображений, которые могут обмануть дискриминатор. Другими словами, генераторы GAN стремятся минимизировать разницу между  и , где  — реальная выборка, а  — сгенерированная с учётом  поддельная выборка. Фактически GAN можно использовать для имитации любого распределения данных (изображения, текста, звука и т. д.).

### ПРИНЦИП РАБОТЫ GAN

Принцип работы GAN можно сравнить с игрой двух игроков: нейронной сети-генератора G и нейронной сети-дискриминатора D. G пытается обмануть D, генерируя реально выглядящие изображения, в то время как D пытается различать настоящие и поддельные изображения. Отсюда и берётся термин «состязательный» в названии модели.


На вход G подаётся простое распределение (случайный шум). Во время обучения генератор постепенно создаёт всё более реалистичные изображения, в то время как дискриминатор всё лучше распознает их. Процесс достигает равновесия, когда дискриминатор больше не может отличать реальные изображения от поддельных. Таким образом, если дискриминатор хорошо обучен и генератору удаётся генерировать реально выглядящие изображения, которые обманывают дискриминатор, то у нас есть хорошая генеративная модель: мы генерируем изображения, которые выглядят как обучающая выборка.

### 3. Генерация изображений с помощью GAN
Итак, вы знаете, что GAN включает в себя две подмодели: генератор и дискриминатор. Оптимизировать их довольно сложно, потому что, как вы можете себе представить, две сети преследуют совершенно противоположные цели: генератор хочет создать что-то максимально реалистичное, в то время как дискриминатор хочет различать сгенерированные материалы.
