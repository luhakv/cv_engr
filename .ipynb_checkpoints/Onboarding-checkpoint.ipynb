### Добро пожаловать в трек CV!


️ ️→ Теперь, когда вы сделали выбор трека, мы можем приступить к освоению направления Computer Vision.

Вы уже многое узнали о сфере Data Science и освоили умения, которых достаточно, чтобы претендовать на позицию дата-сайентиста. За следующие семь месяцев вы отточите навыки в отдельной сфере DS — в сфере компьютерного зрения.

Вы научитесь решать такие задачи CV, как классификация изображений, сегментация и детекция, генерация и стилизация картинок, восстановление и повышение качества фотографий, а также работа с видеорядом. ### ЧТО НЕОБХОДИМО ЗНАТЬ

За прошедший год вы освоили целый пул компетенций Data Science, и при прохождении CV-трека вам предстоит часто обращаться к этим знаниям.

- Python. Все модели, которые мы будем реализовывать, будут написаны на языке Python.

- Линейная алгебра, математический анализ. Вам должно быть комфортно брать производные и понимать операции и обозначения векторных матриц. В ходе курса мы еще повторим и отработаем эти навыки.

- Базовая теория вероятности и статистика. Вам нужно знать основы вероятностей, гауссовских распределений, среднего, стандартного отклонения и так далее.

- Знакомство с принципами и алгоритмами машинного обучения. Нужно быть знакомым с составляющими ML-моделей: функцией потерь, процессом обучения, оптимизацией, регуляризацией, а также с такими моделями, как регрессия, классификация, кластеризация и прочими.

- Общее представление о глубоком обучении. Большинство современных CV-систем построены на нейронных сетях. В качестве основного фреймворка мы будем использовать PyTorch.

- Знание английского на уровне B1-B2. Это поможет в чтении актуальных статей и материалов, которые мы включили как в дополнительные, так и в обязательные материалы курса.

* Напомним, что материалы прошедшего года обучения остаются доступными вам, а потому вы можете в любое время обратиться к ним для повторения.

### ЧТО ВЫ ПРИОБРЕТЁТЕ В КОНЦЕ КУРСА

Вы не только научитесь решать все базовые задачи с сфере CV, но и приобретёте навыки, которые будут выгодно выделять вас среди других начинающих CV-специалистов: знание реального флоу работы с моделями CV, актуальных подходов и продвинутых инструментов, необходимых для создания CV-сервисов.

Отдельное внимание мы уделим подготовке к поиску работы в сфере CV, отбору и прохождению собеседований.

### Как проходит обучение?

⛏ Для наc важное значение имеет достижение той профессиональной цели, которую вы сформулировали. Поэтому методика, на которой основан трек, направлена на максимальное содействие этому.

### ПРОЕКТНАЯ РАБОТА

Трек ориентирован на решение реальных практических задач. В рамках трека вам необходимо будет разобрать и реализовать несколько проектов, поучаствовать в хакатоне, а также реализовать индивидуальный дипломный проект. Это позволит вам пополнить портфолио в области CV.

### ДИПЛОМНЫЙ ПРОЕКТ

Индивидуальный проект даст вам возможность исследовать отдельную тему более глубоко и определиться с областью дальнейшего развития внутри сферы компьютерного зрения. Работа над индивидуальным проектом будет проходить во второй половине курса под руководством опытного ментора — представителя индустрии.

### ФИНАЛЬНАЯ АТТЕСТАЦИЯ И РЕПЕТИЦИОННОЕ СОБЕСЕДОВАНИЕ

В конце курса, помимо сдачи дипломного проекта, вы попробуете себя в репетиционном собеседовании. Так вы сможете подготовиться к отбору на реальные вакансии, которые вас ждут.

### ИТОГОВАЯ ОЦЕНКА

Ваша оценка складывается из следующих составляющих:

35 %	Тестовые и итоговые задания с автоматической проверкой внутри отдельных модулей  
20 %	Проекты и хакатоны по тематическим блокам с проверкой от эксперта  
20 %	Финальная аттестация и прохождение репетиционного собеседования  
25 %	Выполнение финального дипломного проекта  

Для получения диплома вам необходимо перейти порог в 70 % от максимальной оценки курса.
