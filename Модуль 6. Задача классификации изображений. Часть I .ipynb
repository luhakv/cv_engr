### 2. Задача классификации изображений

⛏ Перед погружением в дебри классических подходов и современных архитектур свёрточных нейронных сетей давайте выясним, в чём вообще состоит задача классификации изображений и какие у неё особенности.

Смысл классификации в том, что по входной картинке мы пытаемся определить, что за объект на ней изображён. В этом случае мы обычно полагаемся на то, что на картинке действительно есть один доминантный объект, будь то кошка, собака или автомобиль. Ещё один вариант — мы пытаемся отнести картинку к одной из заранее известных категорий («утро», «пейзажи», «импрессионизм», «запрещённый контент» и так далее).

С первого взгляда задача кажется простой, так как человек приспособился решать её сам для огромного числа категорий: выделяют до 3 000 основных существительных в 10 крупных подкатегориях, в крупных известных датасетах с изображениями типа ImageNet определено 21 000 классов, иногда выделяют и больше классов — до 100 000).

Проблема в том, что люди классифицируют неосознанно. Нам же нужно научить компьютер понимать, что на картинке изображён кот. Это задача не из лёгких, учитывая, что компьютеру нужно будет сделать вывод по, например, 1 440 000 числовых значений (это число соответствует цветной картинке размера 800x600).



И это картинка только одного конкретного кота — в реальности изображения могут сильно отличаться в освещённости, ракурсах, масштабах, да и коты есть разные!

### Задание 2.2

С какой принципиальной проблемой мы сталкиваемся в задаче классификации?
- Большое разнообразие изображений внутри класса 



→ Это так называемая проблема **семантического разрыва** из-за непонимания информации, которая заключена в числовых данных — мы видим, что изображено на картинке, но непонятно, как это вывести из числового представления, изменчивого из-за упомянутых обстоятельств.

Нам нужны такие алгоритмы классификации изображений, которые были бы устойчивы ко всем рассмотренным изменениям. О **классических и современных нейросетевых подходах** к этой задаче мы и поговорим в этом и последующем модулях.
