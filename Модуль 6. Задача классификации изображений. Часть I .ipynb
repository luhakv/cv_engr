### Модуль 6. Задача классификации изображений. Часть I 

### 1. Что вас ждёт в этом модуле?

→ Теперь вы больше знаете о глубоком обучении, а также владеете инструментами, необходимыми для CV-специалиста. В этом и следующем модулях вы погрузитесь в одну из задач компьютерного зрения — задачу классификации.

### В ЭТОМ МОДУЛЕ МЫ РАЗБЕРЁМ:

- какие особенности есть у задачи классификации;
- почему свёрточные нейросети хороши для распознавания изображений;
- как эволюционировали архитектуры для задач классификации;
- какие метрики используются для оценки моделей классификации.

### ВЫ НАУЧИТЕСЬ:

- писать собственные модели для распознавания изображений;
- использовать предобученные сети для классификации.
