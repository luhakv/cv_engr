### Модуль 8. Перенос обучения в CV 

### 1. Что вас ждёт в этом модуле?

⛏ В предыдущих модулях мы повторили основы математики и статистики, необходимые для данного курса. Теперь вам будет проще углубиться в машинное обучение и понять, как работают те или иные алгоритмы. Благодаря накопленным знаниям вы с лёгкостью перейдёте к изучению новых тем.

Традиционно алгоритмы машинного и глубокого обучения разрабатывались для изолированной работы. Они обучены решать конкретные задачи, а после изменения признакового описания модели необходимо перестраивать с нуля.

**Идея переноса обучения** — это преодоление изолированной парадигмы обучения и использование знаний, полученных для одной задачи, к решению связанных с ней задач.

В этом модуле мы подробно рассмотрим концепцию переноса обучения, его преимущества и подводные камни, а также продемонстрируем некоторые практические примеры.

- что такое перенос обучения;

- какие методы переноса обучения существуют и от чего зависит их выбор;

- чём состоят преимущества и риски применения переноса на реальных задачах.
